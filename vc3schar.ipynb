```python
from dlogprocess.char.char_dlog import CharDlog
from collections import Counter
import pandas as pd
```

```python
vc3schar = CharDlog(r'./Data/Char/Room_30X.txt', '25C')
```

```python
data = vc3schar.gen_df()
```

```python
tests = list(Counter(data.Test).keys())
```

```python
pins = list(Counter(data.Pin).keys())
```

```python
data = data.convert_objects(convert_numeric=True)
```

```python
data.dtypes
```

```python
def get_summary(vddmin, vddmax, data):
    _df = []
    for x in tests:
        for y in pins:
            if not(data[(data.Test == x) & (data.Pin == y) & (data.VDD > vddmin) & (data.VDD < vddmax)].Data.empty):
    #             print(data[(data.Test == x) & (data.Pin == y)].Data.describe())
                _dfbuffer = list(data[(data.Test == x) & (data.Pin == y) & (data.VDD > vddmin) & (data.VDD < vddmax)].Data.describe())[1:]
                _dfbuffer.insert(0, y)
                _dfbuffer.insert(0, x)
                _df.append(_dfbuffer)
    df = pd.DataFrame(_df, columns = ['Test', 'Pin', 'mean', 'std', 'min', '_25p', '_50p', '_75p', 'max'])
    return df
```

```python
sum33 = get_summary(3.3*0.85, 3.3*1.15, data)
sum33.to_csv('sum33.csv')
sum25 = get_summary(2.5*0.85, 2.5*1.15, data)
sum25.to_csv('sum25.csv')
sum18 = get_summary(1.8*0.85, 1.8*1.15, data)
sum18.to_csv('sum18.csv')
```
